```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',100)
```

```python
df =pd.read_excel('Superstore_USA.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Product Base Margin']=df['Product Base Margin'].fillna(df['Product Base Margin'].mean())
```

```python
df.isnull().sum()
```

```python
df.head()
```

# Order Priority

```python
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority'].replace('Critical ','Critical',inplace=True)
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(data=df,x='Order Priority',hue='Order Priority')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title("Count of Order Priority")
plt.show()
```

# Shipping Mode

```python
sm=df['Ship Mode'].value_counts()
```

```python
ax=sns.countplot(data=df,x='Ship Mode',hue='Ship Mode' )
ax.set_title('Shipping Mode Count')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.pie(sm.values,labels=sm.index,autopct='%0.2f%%',startangle=90 )
plt.legend()
plt.show()
```

# Product Category

```python
ax=sns.countplot(data=df,x='Product Category',hue='Product Category' )
ax.set_title('Product Category Count')
for bars in ax.containers:
    ax.bar_label(bars)
```
